```julia
Ks = 1:20; Ks_ext = 1:30
bell_numbers = [ bell_number(k) for k in Ks ]
bounds = [ (0.792*k/log(1+k))^k for k in Ks_ext ]

using PyPlot
plot(Ks_ext, bounds, "rv-", label="Bound")
plot(Ks, bell_numbers, "b.-", label="Bell number")
yscale(:log)
legend(loc="best")
xlabel("Number of BSs"); ylabel("Number of partitions")
```
