```julia
Ks = 1:30
bell_numbers = [ bell_number(k) for k in Ks ]
bounds = [ (0.792*k/log(1+k))^k for k in Ks ]

max_cluster_size = 4 # largest coalition size which is IA feasible
restricted_bounds = zeros(BigInt, length(Ks))
for k in Ks[1:end]
    stirlings = collect(Stirling2NumberIterator(k))
    restricted_bounds[k] = sum(stirlings[iceil(k/max_cluster_size):end])
end
```

### Compare Bell number with restricted complexity number

```julia
using PyPlot
plot(Ks, bounds, "rv-", label="Bound")
plot(Ks, bell_numbers, "b.", label="Bell number")
plot(Ks, restricted_bounds, "g--", label="Exhaustive search complexity")
yscale(:log)
legend(loc="best")
xlabel("Number of BSs"); ylabel("Number of partitions")
```
